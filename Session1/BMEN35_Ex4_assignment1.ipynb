### 1. Generate x-values (1000 values between 0 and 5)

```python
import numpy as np
X1 = np.random.rand(1000)*5  #Random number between 0 to 5
```

### 2. Generate y-values according to some function (y = x, y = e^x, y = x^2 or similar) that is defined through all x-values

```python
#y = X1
#y= np.exp(X1)
y = X1**2
```

### 3. Take 100 random samples from the y array. (tip: np.random.randint(x,y,z) generates z number of integer values between x and y). Compute the mean of these 100 random samples.

```python
y_100_samples = y[np.random.randint(0,1000,100)]
np.mean(y_100_samples)
```

### 4. Repeat step 3, 50000 times and store the means in a variable (clt)

```python
q = np.array([y[np.random.randint(0,1000,100)] for i in range(0, 5000)]) #we make a matrix of all values.
clt = np.mean(q,axis = 1) #taking the mean of each row
```

### 5. Plot a histogram of the clt variable

```python
import matplotlib.pyplot as plt 
plt.hist(clt, rwidth=0.95)
plt.xlabel("y_mean")
plt.ylabel("antal")
plt.show()
```

### 6. Go back and change the function in step 2 to something else and run the script again.

```python
y= np.exp(X1)
q = np.array([y[np.random.randint(0,1000,100)] for i in range(0, 5000)])
clt = np.mean(q,axis = 1)
plt.hist(clt, rwidth=0.95)
plt.xlabel("y_mean")
plt.ylabel("Count")
plt.show()
```
